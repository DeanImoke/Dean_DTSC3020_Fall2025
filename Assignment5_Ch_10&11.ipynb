```python
import re
from pathlib import Path

def is_valid_email(email):
    """Validates an email address using a regex."""
    if not email:
        return False
    email = email.strip()
    pattern = r"^[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}$"
    return re.fullmatch(pattern, email) is not None

def normalize_phone(phone):
    """Normalizes a phone number by removing non-digits and keeping the last 10."""
    if not phone:
        return ""
    digits = re.sub(r"\D", "", phone)
    if len(digits) >= 10:
        return digits[-10:]
    return ""

cleaned_contacts = []
seen_emails = set()

path = Path("contacts_raw.txt")

try:
    with open(path, "r", encoding="utf-8") as f:
        for line in f:
            line = line.strip()
            if not line:
                continue

            # Simple parsing based on common patterns
            name = ""
            email = ""
            phone = ""

            # Pattern 1: Name <email> , phone
            match = re.match(r"^(.*?)\s*<(.*?)>\s*,\s*(.*)$", line)
            if match:
                name = match.group(1).strip().strip('"')
                email = match.group(2).strip()
                phone = match.group(3).strip()
            else:
                # Pattern 2: Name , email , phone
                parts = [p.strip() for p in line.split(',')]
                if len(parts) >= 3:
                    name = parts[0].strip().strip('"')
                    email = parts[1].strip()
                    phone = parts[2].strip()
                elif len(parts) == 2:
                     # Assuming Name, email
                     name = parts[0].strip().strip('"')
                     email = parts[1].strip()


            if is_valid_email(email):
                normalized_phone = normalize_phone(phone)
                casefolded_email = email.casefold()

                if casefolded_email not in seen_emails:
                    cleaned_contacts.append({
                        "name": name,
                        "email": email,
                        "phone": normalized_phone
                    })
                    seen_emails.add(casefolded_email)

except FileNotFoundError:
    print("Error: The file 'contacts_raw.txt' was not found.")

# Write to CSV
output_path = Path("contacts_clean.csv")
try:
    with open(output_path, "w", encoding="utf-8") as f:
        f.write("name,email,phone\n")
        for contact in cleaned_contacts:
            f.write(f'"{contact["name"]}","{contact["email"]}","{contact["phone"]}"\n')
except IOError:
    print(f"Error: Could not write to file {output_path}")
```

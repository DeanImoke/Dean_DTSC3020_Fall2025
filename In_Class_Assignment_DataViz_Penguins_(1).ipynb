## 0) Setup

```python
# Run it first.
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.set_theme(style='whitegrid')
penguins = sns.load_dataset('penguins')
penguins = penguins.dropna()
penguins.head()  # preview
```

## 1–3) Scatter plot with color by species
Create a scatter plot with:
- **x =** `bill_length_mm`
- **y =** `bill_depth_mm`
- Color points by **species**

Then add a short comment (2–3 sentences) describing the pattern(s) you see.

```python
# TODO: Create the scatter plot here
sns.scatterplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', hue='species')
plt.title('Bill Length vs. Bill Depth')
plt.show()



```

**Your interpretation (2–3 sentences):**

I see that Adelies have shorter bill length and more bill depth. Chinstraps and Getoos have similar bill length, but Chinstraps have more bill depth. The longer the bill, the more depth.

## 4) Histogram with two different bin sizes
Plot a histogram of `bill_length_mm` **twice**:
- Once with **10 bins**
- Once with **30 bins**

Then write 1–2 sentences comparing what changes when you change the number of bins.

```python
# TODO: Histogram with 10 bins
sns.histplot(data=penguins, x='bill_length_mm', bins=10)
plt.title('Bill Length Histogram (10 bins)')
plt.show()

```

```python
# TODO: Histogram with 30 bins
sns.histplot(data=penguins, x='bill_length_mm', bins=30)
plt.title('Bill Length Histogram (10 bins)')
plt.show()
```

**Your comparison (1–2 sentences):**
When you have more bins, the results are more accurate. Most bills are between 35 and 53 millimeters.

## 5) Boxplot of body mass by species
Create a **boxplot** showing `body_mass_g` grouped by `species`.

```python
# TODO: Boxplot of body mass by species
sns.boxplot(data=penguins, x='species', y='body_mass_g')
plt.title('Body Mass by Species')
plt.show()
```
